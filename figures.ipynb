```python
import matplotlib.pyplot as plt


Bitrates = [3.31, 3.99, 4.88, 6.13, 7.73, 9.44, 11.23, 12.81, 13.10, 13.13, 13.23, 13.43, 13.89, 20.08]  
D2_PSNR = [48.23, 54.34, 60.37, 66.54, 72.52, 78.56, 84.58, 90.57, 96.58, 102.63, 108.64, 114.65, 120.71, 126.24] 


ours_Bitrates = [1.79, 2.78, 4.34, 6.12, 7.92, 9.72, 11.51, 13.31, 15.11, 16.91]    
ours_D2_PSNR = [66.15, 72.18, 78.18, 84.10, 89.61, 94.17, 96.78, 97.80, 98.11, 98.19] 

KDDI_Bitrates = [2.18, 3.90, 9.12, 13.97, 21.25]    
KDDI_D2_PSNR = [70.79, 71.15, 77.26, 80.01, 80.75] 


# Plotting
plt.figure(figsize=(12, 12))
plt.plot(Bitrates, D2_PSNR, marker='o', label='Draco', markersize=14)
plt.plot(ours_Bitrates, ours_D2_PSNR, marker='s', label='Ours, GoF = 5', markersize=14)
plt.plot(KDDI_Bitrates, KDDI_D2_PSNR, marker='^', label='KDDI', markersize=14)
#plt.title('Dancer - fps: 30')
plt.xlabel('Bitrate (Mbps)', fontsize=32)
plt.ylabel('D2-PSNR', fontsize=32)
plt.grid(True)
plt.legend(fontsize=30, loc = 'lower right')  # Show legend with labels
plt.xlim(0, 25)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.tight_layout()

# Display the plot
plt.show()
```
